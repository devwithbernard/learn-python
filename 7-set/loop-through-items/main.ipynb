```python
"""
Loop through set items
"""

def main() -> None:
    fruits: set[str] = {'Apple', 'Banana', 'Orange', 'Cherry'}
    
    def loop_through_items(sequence: set[str]) -> None:
        for index, fruit in enumerate(sequence):
            print(f"{index} => {fruit}")
            index += 1

    loop_through_items(fruits)
    
if __name__ == '__main__':
    main()
```
