```python
from typing import List, Dict, Any
import requests

def get_users(url: str) -> List[Dict]:
    try:
        response = requests.get(url)
        return response.json()
    except requests.HTTPError as http_error:
        print(f"{http_error.__str__()}")
    except requests.RequestException as r_exception:
        print(f"{r_exception.__str__()}")
    except Exception as e:
        print(f"Error occurs: {e.__str__()}")

def format_users(users: List[Dict]) -> None:
    if users:
        for user in users:
            print(f"""
            User n°{user['id']}
                id => {user['id']}
                name => {user['name']}
                email => {user['email']}
            """)
        
def main() -> None:
    users: List[Dict] = get_users('https://jsonplaceholder.typicode.com/users')
    format_users(users)
    
if __name__ == '__main__':
    main()
```
